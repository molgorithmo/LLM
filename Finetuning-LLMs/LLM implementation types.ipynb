# Chatbot

```python
from langchain_openai import ChatOpenAI
from langchain.prompts.chat import (
    ChatPromptTemplate,
    SystemMessagePromptTemplate,
    HumanMessagePromptTemplate
)
```

```python
chat_llm = ChatOpenAI(model='gpt-3.5-turbo', temperature=0)
```

```python
#help(chat_llm) # Very useful
```

```python
template = "You are a beautiful assistant that will help users find information about movies."
system_message_prompt = SystemMessagePromptTemplate.from_template(template)

human_template = "Find information about the movie {movie_title}."
human_message_prompt = HumanMessagePromptTemplate.from_template(human_template)
```

```python
chat_prompt = ChatPromptTemplate.from_messages([system_message_prompt, human_message_prompt])
```

```python
response = chat_llm(chat_prompt.format_prompt(movie_title="The Matrix").to_messages())
```

```python
print(response.content)
```

Using the `to_messages` object in LangChain allows you to convert the formatted value of a chat prompt template into a list of message objects. This is useful when working with chat models, as it provides a structured way to manage the conversation and ensures that the chat model can understand the context and roles of the messages.



```python
chat_prompt.format_prompt(movie_title="The Matrix").to_messages()
```

```python
from langchain import OpenAI, PromptTemplate
from langchain.chains.summarize import load_summarize_chain
from langchain.document_loaders import PyPDFLoader
```

```python
llm = OpenAI(model='gpt-3.5-turbo-instruct', temperature=0)
```

```python
summarize_chain = load_summarize_chain(llm)
```

```python
document_loader = PyPDFLoader('/content/drive/MyDrive/Active Loop RAG/Resume_Mohit_Sharma.pdf')
```

```python
document = document_loader.load()
```

```python
summary = summarize_chain.invoke(document)
print(summary['output_text'])
```
